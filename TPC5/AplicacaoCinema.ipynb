```python
#Cinema = [Sala]
#Sala = [nlugares, Vendidos, filme]
#nlugares = Int
#filme = String 
#Vendidos = [Int]

def listarCinema(cinema):
    for sala in cinema:
        nome = sala[2]
        print(f"esta em exibicao o filme {nome}")

def disponivel(cinema, nome, lugar):
    cond = False
    for sala in cinema:
        if sala[2] == nome:
            if lugar not in sala[1]:
                cond = True           
    return cond

def vendeBilhete(cinema,filme,lugar):
    for sala in cinema:
        if sala[2]==filme:
            sala[1].append(lugar)
    return cinema
    
def listarDisponibilidades(cinema):
    lugares_disponiveis = 0
    for sala in cinema:
        nlugares, vendidos, filme = sala
        lugares_disponiveis = nlugares - len(vendidos)
        print(f"o filme em exibicao é {filme} e os lugares disponiveis sao {lugares_disponiveis}")
    return 

def existeSala(cinema, filme):
    cond = False
    for sala in cinema:
        if sala[2] == filme:
            cond = True
    return cond
        
def inserirSala(cinema,filme):
    if not existeSala(cinema,filme):
        cinema.append(filme)
    return cinema
```

```python
#testar funções
sala1 = (150, [], "Twilight")
sala2 = (200, [1,3], "Hannibal")
cinema1 = []
cinema1.append(sala1)
cinema1.append(sala2)

print(listarCinema(cinema1))
print(disponivel(cinema1,"Twilight",1))
print(disponivel(cinema1,"Hannibal",1))
print(vendeBilhete(cinema1,"Hannibal",2))
print(listarDisponibilidades(cinema1))
sala5= (300, [1,3,4], "Divertidamente")
cinema1 = inserirSala(cinema1, sala5)
print(listarCinema(cinema1))
print(listarDisponibilidades(cinema1))
```
