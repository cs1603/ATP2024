```python
lista1 = [1, 2, 3, 4, 5]
lista2 = [4, 5, 6, 7, 8]  
ncomuns = [x for x in (lista1 + lista2) if x not in lista1 or x not in lista2] #significa que x vai ser incluido se estiver ausente numa das listas originais
print(ncomuns)
# Resultado esperado: [1,2,3,7,8] [1,2,3,4,5,4,5,6,7,8]

texto = """Vivia há já não poucos anos algures num concelho do Ribatejo 
    um pequeno lavrador e negociante de gado chamado Manuel Peres Vigário"""
lista = [pal for pal in texto.split() if len(pal)>3]
print(lista)
# Resultado esperado: ['Vivia', 'poucos', 'anos', 'algures', 'concelho', ...]


lista = ['anaconda', 'burro', 'cavalo', 'macaco']
listaRes =[(x+1, lista[x]) for x in range (len(lista))]
print(listaRes)
# Resultado esperado: [(1,'anaconda'), (2,'burro'), (3,'cavalo'), (4,'macaco')]

def strCount(s, subs):
    contador = 0
    i = 0
    while i <= len(s)-len(subs):
        if s[i: i+len(subs)]== subs: # Verifica se há uma ocorrência de 'subs' a partir de 'i'
            contador = contador + 1
            i = i + len(subs)
        else:
            i = i + 1
    return contador

print(strCount("catcowcat", "cat")) # --> 2
print(strCount("catcowcat", "cow")) # --> 1
print(strCount("catcowcat", "dog")) # --> 0


def produtoM3(lista):
    nova_lista = sorted(lista)
    for x in nova_lista:
        produto= (nova_lista[0]*nova_lista[1]*nova_lista[2])
    return produto
            

print(produtoM3([12,3,7,10,12,8,9]))
# Resultado esperado: 168 = 3 * 7 * 8

# Input: 38
# Output: 2
# Explicação: 3 + 8 = 11, 1 + 1 = 2.

# Input: 777
# Output: 3
# Explicação: 7 + 7 + 7 = 21, 2 + 1 = 3.

def reduxInt(n):
    while n >= 10:  # Continua enquanto n tiver mais de um dígito
        soma = 0
        for digit in str(n):  # Converte o número para string para iterar sobre os dígitos
            soma += int(digit)  # Converte cada dígito de volta para inteiro e soma
        n = soma  # Atualiza n com a nova soma
    return n

print(reduxInt(38))
print(reduxInt(777))

# Invocação: indexOf("Hoje está um belo dia de sol!", "belo")
# Resultado: 13

# Invocação: indexOf("Hoje está um belo dia de sol!", "chuva")
# Resultado: -1

def myIndexOf(s1, s2):
    res = -1
    i = 0
    while i <= len(s1) - len(s2):
        if s1[i : i + len(s2)] == s2:
            res = i
            i = i + len(s2)
        else:
            i = i + 1
    return res

print(myIndexOf("Hoje está um belo dia de sol!", "belo"))
print(myIndexOf("Hoje está um belo dia de sol!", "chuva"))






MyFaceBook = [{
        'id': 'p1', 
        'conteudo': 'A tarefa de avaliação é talvez a mais ingrata das tarefas que um professor tem de realizar...', 
        'autor': 'jcr', 
        'dataCriacao': '2023-07-20', 
        'comentarios': [
            {
                'comentario': 'Completamente de acordo...',
                'autor': 'prh'
            },
            {
                'comentario': 'Mas há quem goste...',
                'autor': 'jj'
            }]}, 
            {'id': 'p2', 
        'conteudo': 'A tarefa de avaliação é talvez a mais ingrata das tarefas que um professor tem de realizar...', 
        'autor': 'ana', 
        'dataCriacao': '2023-07-20', 
        'comentarios': [
            {
                'comentario': 'Completamente de acordo...',
                'autor': 'prh'
            },
            {
                'comentario': 'Mas há quem goste...',
                'autor': 'jj'
            }]
            }]


def quantosPost(redeSocial):
    return len(redeSocial)

def postsAutor(redeSocial, autor):
    lista_post = []
    for post in redeSocial:
        if post["autor"] == autor:
            lista_post.append(post)
    return lista_post

postsAutor(MyFaceBook, "jcr")


def autores(redeSocial):
    autores = []
    for post in redeSocial:
        autores.append(post["autor"])     
    return sorted(autores)

print(autores(MyFaceBook))



def insPost(redeSocial, conteudo, autor, dataCriacao, comentarios):
    id = "p" + str(len(redeSocial)+1)
    novoPost = {
        "id": id,
        "conteudo": conteudo,
        "autor": autor,
        "dataCriacao": dataCriacao,
        "comentarios": comentarios
    }
    redeSocial.append(novoPost)
    return redeSocial

insPost(MyFaceBook, "Novo conteúdo", "Novo autor", "2023-10-30",[{}])


def remPost(redeSocial, id):
    novaRede = redeSocial
    for post in novaRede:
        if post["id"] == id:
            novaRede.remove(post) 
    return novaRede

remPost(MyFaceBook,"p3")

def postsPorAutor(redeSocial):
    res = {}
    for post in redeSocial:
        if post["autor"] in res.keys():
            res[post["autor"]] = res[post["autor"]] + 1
        else:
            res[post["autor"]] = 1
    return res

postsPorAutor(MyFaceBook)

def comentadoPor(redeSocial, autor):
    comentado = []
    for post in redeSocial:
        for comentarios in post["comentarios"]:
            if comentarios["autor"]== autor:
                comentado.append(post)
    return comentado

comentadoPor(MyFaceBook,"jj")
```
