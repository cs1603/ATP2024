```python
#aluno = (nome, id, [notaTPC, notaProj, notaTeste])

#turma = [aluno]

def criar_aluno():
    nome = input("Digite o nome do aluno: ")
    aluno_id = input("Digite o ID do aluno: ")
    notaTPC = float(input("Digite a nota do TPC: "))
    notaProj = float(input("Digite a nota do projeto: "))
    notaTeste = float(input("Digite a nota do teste: "))
    notas = [notaTPC, notaProj, notaTeste]
    aluno = (nome, aluno_id, notas)
    return aluno

# Função para criar uma turma (lista de alunos)
def criar_turma():
    turma1 = []
    while True:
        aluno = criar_aluno()
        turma1.append(aluno)
        return turma1

def aluno_naoexiste(turma, nome, id,):
    cond = True
    for al in turma:
        if id== al[1]:
            cond = False
        else:
            cond 
    return cond
        
def inserir_aluno(turma, nome, id, notaTPC, notaProj, notaTeste):
    if aluno_naoexiste(turma, nome, id):
        turma.append((nome, id, [notaTPC, notaProj, notaTeste]))
    return turma

def listar_turma(turma):
    for aluno in turma:
            print(f"Nome: {aluno[0]}, ID: {aluno[1]}, Notas: {aluno[2]}")
    return turma

def consultar_aluno(turma, id):
    consulta = []
    for aluno in turma:
        if aluno[1] == id:
            consulta.append(aluno)
    return consulta
            
        
def guardar_turma(ficheiro,turma):
    file = open(ficheiro, "w")
    for aluno in turma:
        file.write(str(aluno))
        file.write("\n")
    file.close()

def carregar_turma(ficheiro):
    file = open(ficheiro, "r")
    i=1 #serve para dar o numero da linha 
    for linha in file:
        print(str(i)+ ": :" + linha, end='')
        i = i + 1
    file.close()
```
