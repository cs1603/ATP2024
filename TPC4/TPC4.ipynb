```python
import random

def lista_aleatoria():
    return [random.randint(1, 101) for _ in range(7)]

def ler_lista():
    lista = []
    n = int(input("quantos numeros quer introduzir?"))
    while 0 < n:
        y = int(input("escreva os numeros"))
        lista.append(y)
        n = n - 1
    return lista

def soma_lista(lista):
    soma = 0
    for x in lista:
        soma = soma + x
    return soma

def media_lista(lista):
    g = len(lista)
    soma = 0
    for x in lista:
        soma = soma + x
    media = soma/g
    return media

def maior_elemento(lista):
    maior = lista[0]
    i = 0
    while i < len(lista):
        if lista[i] > maior:
            maior = lista[i]
        i = i + 1
    return maior

def menor_elemento(lista):
    menor = lista[0]
    i = 0
    while i< len(lista):
        if lista[i] < menor:
            menor = lista[i]
        i = i + 1
    return menor

def ordenada_crescente(lista):
    lista_ord = sorted(lista)

    if lista == lista_ord:
        print("Sim, esta ordenada crescente")

    else:
        print("Nao esta ordenada crescente")
    
def ordenada_decrescente(lista):
    lista_decres = sorted(lista, reverse = True)
    if lista == lista_decres:
        print("Sim, esta ordenada decrescente ")
    else:
        print("Nao esta ordenada decrescente")

def procura_elemento(lista):
    elemento = int(input("Digite o elemento a procurar: "))
    if elemento in lista:
        print( lista.index(elemento))
    else:
        print("-1")






def mostrar_menu():
    print("\nMenu:")
    print("(1) Criar Lista")
    print("(2) Ler Lista")
    print("(3) Soma")
    print("(4) Média")
    print("(5) Maior")
    print("(6) Menor")
    print("(7) Esta Ordenada por ordem crescente")
    print("(8) Esta Ordenada por ordem decrescente")
    print("(9) Procura um elemento")
    print("(0) Sair")

def main():
    lista = []
    opcao = ""
    while opcao != "0" :
        mostrar_menu()
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            lista = lista_aleatoria()
            print("Lista criada:", lista)
        elif opcao == "2":
            lista = ler_lista()
            print("Lista lida:", lista)
        elif opcao == "3":
            print("Soma dos elementos:", soma_lista(lista))
        elif opcao == "4":
            print("Média dos elementos:", media_lista(lista))
        elif opcao == "5":
            print("Maior elemento:", maior_elemento(lista))
        elif opcao == "6":
            print("Menor elemento:", menor_elemento(lista))
        elif opcao == "7":
            print("Esta ordenada por ordem crescente?",ordenada_crescente(lista))
        elif opcao == "8":
            print("Esta ordenada por ordem decrescente?",ordenada_decrescente(lista))
        elif opcao == "9":
            posicao = procura_elemento(lista)
            print(posicao)
            
        elif opcao == "0":
            print("Saindo... Lista atual:", lista)
          
    
        
        

if __name__ == "__main__":
    main()
```
